```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
ghana = data[(data['home_team'] == 'Ghana') | (data['away_team'] == 'Ghana')]
```

```python
ghana.head()
```

```python
ghana_l20 = ghana[-20:]
ghana_l20
```

```python
ghana_win = 0
ghana_lose = 0
ghana_draw = 0
for i,j,k,l in zip(ghana_l20['home_team'], ghana_l20['away_team'], ghana_l20['home_score'], ghana_l20['away_score']):
    if ((i == 'Ghana') and (k>l)) or ((j == 'Ghana') and (k<l)):
        ghana_win +=1
    elif ((i == 'Ghana') and (k<l)) or ((j == 'Ghana') and (k>l)):
        ghana_lose +=1
    else:
        ghana_draw += 1
    
print(ghana_win, ghana_lose, ghana_draw)
```

```python
comoros = data[(data['home_team'] == 'Comoros') | (data['away_team'] == 'Comoros')]
```

```python
comoros.head()
```

```python
comoros_l20 = comoros[-20:]
comoros_l20
```

```python
comoros_win = 0
comoros_lose = 0
comoros_draw = 0
for i,j,k,l in zip(comoros_l20['home_team'], comoros_l20['away_team'], comoros_l20['home_score'], comoros_l20['away_score']):
    if ((i == 'Comoros') and (k>l)) or ((j == 'Comoros') and (k<l)):
        comoros_win +=1
    elif ((i == 'Comoros') and (k<l)) or ((j == 'Comoros') and (k>l)):
        comoros_lose +=1
    else:
        comoros_draw += 1
    
print(comoros_win, comoros_lose, comoros_draw)
```

```python
labels = ['WIN', 'LOSE', 'DRAW']

gha = [ghana_win, ghana_lose, ghana_draw]
com = [comoros_win, comoros_lose, comoros_draw]

width = 0.25

x = np.arange(0, 3)

fig, ax = plt.subplots(figsize=(10,10))

y = list(range(0, 16, 2))

rects1 = ax.bar(x - width/2, gha, width, label='Ghana', color='y')
rects2 = ax.bar(x + width/2, com, width, label='Comoros', color='g')

ax.set_ylabel('Number of Outcomes',fontsize=30, color='r')
ax.set_xlabel("Outcomes", fontsize=30, color='r')
ax.set_title("The last 20 matches of Ghana and Comoros", fontsize=25, color='b')
ax.set_xticklabels(labels, fontsize=15, color='b')
ax.set_yticklabels(y, rotation=45, fontsize=25, color='b')

ax.set_xticks(x)
plt.legend(fontsize=15)
plt.grid(True)
#plt.show()
plt.savefig("ghana_comoros.png")
```
