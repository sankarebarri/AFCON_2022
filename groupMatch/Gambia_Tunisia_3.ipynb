```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
gambia = data[(data['home_team'] == 'Gambia') | (data['away_team'] == 'Gambia')]
```

```python
gambia.head()
```

```python
gambia_l20 = gambia[-20:]
gambia_l20
```

```python
gambia_win = 0
gambia_lose = 0
gambia_draw = 0
for i,j,k,l in zip(gambia_l20['home_team'], gambia_l20['away_team'], gambia_l20['home_score'], gambia_l20['away_score']):
    if ((i == 'Gambia') and (k>l)) or ((j == 'Gambia') and (k<l)):
        gambia_win +=1
    elif ((i == 'Gambia') and (k<l)) or ((j == 'Gambia') and (k>l)):
        gambia_lose +=1
    else:
        gambia_draw += 1
    
print(gambia_win, gambia_draw, gambia_lose)
```

```python
tunisia = data[(data['home_team'] == 'Tunisia') | (data['away_team'] == 'Tunisia')]
```

```python
tunisia.tail()
```

```python
tunisia_l20 = tunisia[-20:]
tunisia_l20
```

```python
tunisia_win = 0
tunisia_lose = 0
tunisia_draw = 0
for i,j,k,l in zip(tunisia_l20['home_team'], tunisia_l20['away_team'], tunisia_l20['home_score'], tunisia_l20['away_score']):
    if ((i == 'Tunisia') and (k>l)) or ((j == 'Tunisia') and (k<l)):
        tunisia_win +=1
    elif ((i == 'Tunisia') and (k<l)) or ((j == 'Tunisia') and (k>l)):
        tunisia_lose +=1
    else:
        tunisia_draw += 1
    
print(tunisia_win, tunisia_draw, tunisia_lose)
```

```python
labels = ['WIN', 'DRAW', 'LOSE']

gam = [gambia_win, gambia_draw, gambia_lose]
tun = [tunisia_win, tunisia_draw, tunisia_lose]


width = 0.25

x = np.arange(0, 3)

fig, ax = plt.subplots(figsize=(10,10))

y = list(range(0, 16, 2))

rects1 = ax.bar(x - width/2, gam, width, label='Gambia', color='b')
rects2 = ax.bar(x + width/2, tun, width, label='Tunisia', color='r')

ax.set_ylabel('Number of Outcomes',fontsize=30, color='y')
ax.set_xlabel("Outcomes", fontsize=30, color='y')
ax.set_title("The last 20 matches of Gambia and Tunisia", fontsize=25, color='g')
ax.set_xticklabels(labels, fontsize=35, color='g')
ax.set_yticklabels(y, rotation=0, fontsize=35, color='g')

ax.set_xticks(x)
plt.legend(fontsize=15)
plt.grid(True)
#plt.show()
plt.savefig("images/gambia_tunisia.png")
```
