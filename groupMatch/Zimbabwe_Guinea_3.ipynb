```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
zimbabwe = data[(data['home_team'] == 'Zimbabwe') | (data['away_team'] == 'Zimbabwe')]
```

```python
zimbabwe.head()
```

```python
zimbabwe_l20 = zimbabwe[-20:]
zimbabwe_l20
```

```python
zimbabwe_win = 0
zimbabwe_lose = 0
zimbabwe_draw = 0
for i,j,k,l in zip(zimbabwe_l20['home_team'], zimbabwe_l20['away_team'], zimbabwe_l20['home_score'], zimbabwe_l20['away_score']):
    if ((i == 'Zimbabwe') and (k>l)) or ((j == 'Zimbabwe') and (k<l)):
        zimbabwe_win +=1
    elif ((i == 'Zimbabwe') and (k<l)) or ((j == 'Zimbabwe') and (k>l)):
        zimbabwe_lose +=1
    else:
        zimbabwe_draw += 1
    
print(zimbabwe_win, zimbabwe_lose, zimbabwe_draw)
```

```python
guinea = data[(data['home_team'] == 'Guinea') | (data['away_team'] == 'Guinea')]
```

```python
guinea.tail()
```

```python
guinea_l20 = guinea[-20:]
guinea_l20
```

```python
guinea_win = 0
guinea_lose = 0
guinea_draw = 0
for i,j,k,l in zip(guinea_l20['home_team'], guinea_l20['away_team'], guinea_l20['home_score'], guinea_l20['away_score']):
    if ((i == 'Guinea') and (k>l)) or ((j == 'Guinea') and (k<l)):
        guinea_win +=1
    elif ((i == 'Guinea') and (k<l)) or ((j == 'Guinea') and (k>l)):
        guinea_lose +=1
    else:
        guinea_draw += 1
    
print(guinea_win, guinea_lose, guinea_draw)
```

```python
labels = ['WIN', 'LOSE', 'DRAW']

zim = [zimbabwe_win, zimbabwe_lose, zimbabwe_draw]
gui = [guinea_win, guinea_lose, guinea_draw]

width = 0.25

x = np.arange(0, 3)

fig, ax = plt.subplots(figsize=(10,10))

y = list(range(0, 16, 2))

rects1 = ax.bar(x - width/2, zim, width, label='Zimbabwe', color='g')
rects2 = ax.bar(x + width/2, gui, width, label='Guinea', color='y')

ax.set_ylabel('Number of Outcomes',fontsize=30, color='r')
ax.set_xlabel("Outcomes", fontsize=30, color='r')
ax.set_title("The last 20 matches of Zimbabwe and Guinea", fontsize=25, color='b')
ax.set_xticklabels(labels, fontsize=15, color='b')
ax.set_yticklabels(y, rotation=0, fontsize=25, color='b')

ax.set_xticks(x)
plt.legend(fontsize=15)
plt.grid(True)
#plt.show()
#plt.savefig("zimbabwe_guinea.png")
```
