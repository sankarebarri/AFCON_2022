```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
gabon = data[(data['home_team'] == 'Gabon') | (data['away_team'] == 'Gabon')]
```

```python
gabon.head()
```

```python
gabon_l20 = gabon[-20:]
gabon_l20
```

```python
gabon_win = 0
gabon_lose = 0
gabon_draw = 0
for i,j,k,l in zip(gabon_l20['home_team'], gabon_l20['away_team'], gabon_l20['home_score'], gabon_l20['away_score']):
    if ((i == 'Gabon') and (k>l)) or ((j == 'Gabon') and (k<l)):
        gabon_win +=1
    elif ((i == 'Gabon') and (k<l)) or ((j == 'Gabon') and (k>l)):
        gabon_lose +=1
    else:
        gabon_draw += 1
    
print(gabon_win, gabon_lose, gabon_draw)
```

```python
morocco = data[(data['home_team'] == 'Morocco') | (data['away_team'] == 'Morocco')]
```

```python
morocco.tail()
```

```python
morocco_l20 = morocco[-20:]
morocco_l20
```

```python
morocco_win = 0
morocco_lose = 0
morocco_draw = 0
for i,j,k,l in zip(morocco_l20['home_team'], morocco_l20['away_team'], morocco_l20['home_score'], morocco_l20['away_score']):
    if ((i == 'Morocco') and (k>l)) or ((j == 'Morocco') and (k<l)):
        morocco_win +=1
    elif ((i == 'Morocco') and (k<l)) or ((j == 'Morocco') and (k>l)):
        morocco_lose +=1
    else:
        morocco_draw += 1
    
print(morocco_win, morocco_lose, morocco_draw)
```

```python
labels = ['WIN', 'LOSE', 'DRAW']

gab = [gabon_win, gabon_lose, gabon_draw]
mor = [morocco_win, morocco_lose, morocco_draw]

width = 0.25

x = np.arange(0, 3)

fig, ax = plt.subplots(figsize=(10,10))

y = list(range(0, 16, 2))

rects1 = ax.bar(x - width/2, gab, width, label='Gabon', color='y')
rects2 = ax.bar(x + width/2, mor, width, label='Morocco', color='r')

ax.set_ylabel('Number of Outcomes',fontsize=30, color='r')
ax.set_xlabel("Outcomes", fontsize=30, color='r')
ax.set_title("The last 20 matches of Gabon and Morocco", fontsize=25, color='b')
ax.set_xticklabels(labels, fontsize=15, color='b')
ax.set_yticklabels(y, rotation=0, fontsize=25, color='b')

ax.set_xticks(x)
plt.legend(fontsize=15)
plt.grid(True)
#plt.show()
plt.savefig("gabon_morocco.png")
```
