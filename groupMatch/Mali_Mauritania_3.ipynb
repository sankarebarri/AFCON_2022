```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
mali = data[(data['home_team'] == 'Mali') | (data['away_team'] == 'Mali')]
```

```python
mali.head()
```

```python
mali_l20 = mali[-20:]
mali_l20
```

```python
mali_win = 0
mali_lose = 0
mali_draw = 0
for i,j,k,l in zip(mali_l20['home_team'], mali_l20['away_team'], mali_l20['home_score'], mali_l20['away_score']):
    if ((i == 'Mali') and (k>l)) or ((j == 'Mali') and (k<l)):
        mali_win +=1
    elif ((i == 'Mali') and (k<l)) or ((j == 'Mali') and (k>l)):
        mali_lose +=1
    else:
        mali_draw += 1
    
print(mali_win, mali_draw, mali_lose)
```

```python
mauritania = data[(data['home_team'] == 'Mauritania') | (data['away_team'] == 'Mauritania')]
```

```python
mauritania.tail()
```

```python
mauritania_l20 = mauritania[-20:]
mauritania_l20
```

```python
mauritania_win = 0
mauritania_lose = 0
mauritania_draw = 0
for i,j,k,l in zip(mauritania_l20['home_team'], mauritania_l20['away_team'], mauritania_l20['home_score'], mauritania_l20['away_score']):
    if ((i == 'Mauritania') and (k>l)) or ((j == 'Mauritania') and (k<l)):
        mauritania_win +=1
    elif ((i == 'Mauritania') and (k<l)) or ((j == 'Mauritania') and (k>l)):
        mauritania_lose +=1
    else:
        mauritania_draw += 1
    
print(mauritania_win, mauritania_draw, mauritania_lose)
```

```python
labels = ['WIN', 'DRAW', 'LOSE']

mli = [mali_win, mali_draw, mali_lose]
maur = [mauritania_win, mauritania_draw, mauritania_lose]


width = 0.25

x = np.arange(0, 3)

fig, ax = plt.subplots(figsize=(10,10))

y = list(range(0, 14, 2))

rects1 = ax.bar(x - width/2, mli, width, label='Mali', color='yellow')
rects2 = ax.bar(x + width/2, maur, width, label='Mauritania', color='g')

ax.set_ylabel('Number of Outcomes',fontsize=30, color='r')
ax.set_xlabel("Outcomes", fontsize=30, color='r')
ax.set_title("The last 20 matches of Mali and Mauritania", fontsize=25, color='b')
ax.set_xticklabels(labels, fontsize=35, color='b')
ax.set_yticklabels(y, rotation=0, fontsize=35, color='b')

ax.set_xticks(x)
plt.legend(fontsize=15)
plt.grid(True)
#plt.show()
plt.savefig("images/mali_mauritania.png")
```
