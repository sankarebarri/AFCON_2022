```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
egypt = data[(data['home_team'] == 'Egypt') | (data['away_team'] == 'Egypt')]
```

```python
egypt.head()
```

```python
egy_l20 = egypt[-20:]
egy_l20
```

```python
egy_win = 0
egy_lose = 0
egy_draw = 0
for i,j,k,l in zip(egy_l20['home_team'], egy_l20['away_team'], egy_l20['home_score'], egy_l20['away_score']):
    if ((i == 'Egypt') and (k>l)) or ((j == 'Egypt') and (k<l)):
        egy_win +=1
    elif ((i == 'Egypt') and (k<l)) or ((j == 'Egypt') and (k>l)):
        egy_lose +=1
    else:
        egy_draw += 1
    
print(egy_win, egy_lose, egy_draw)
```

```python
sudan = data[(data['home_team'] == 'Sudan') | (data['away_team'] == 'Sudan')]
```

```python
sudan.tail()
```

```python
sud_l20 = sudan[-20:]
sud_l20
```

```python
sud_win = 0
sud_lose = 0
sud_draw = 0
for i,j,k,l in zip(sud_l20['home_team'], sud_l20['away_team'], sud_l20['home_score'], sud_l20['away_score']):
    if ((i == 'Sudan') and (k>l)) or ((j == 'Sudan') and (k<l)):
        sud_win +=1
    elif ((i == 'Sudan') and (k<l)) or ((j == 'Sudan') and (k>l)):
        sud_lose +=1
    else:
        sud_draw += 1
    
print(sud_win, sud_lose, sud_draw)
```

```python
labels = ['WIN', 'LOSE', 'DRAW']

egy = [egy_win, egy_lose, egy_draw]
sud = [sud_win, sud_lose, sud_draw]

width = 0.25

x = np.arange(0, 3)

fig, ax = plt.subplots(figsize=(10,10))

y = list(range(0, 16, 2))

rects1 = ax.bar(x - width/2, egy, width, label='Egypt', color='k')
rects2 = ax.bar(x + width/2, sud, width, label='Sudan', color='g')

ax.set_ylabel('Number of Outcomes',fontsize=30, color='r')
ax.set_xlabel("Outcomes", fontsize=30, color='r')
ax.set_title("The last 20 matches of Egypt and Sudan", fontsize=25, color='b')
ax.set_xticklabels(labels, fontsize=15, color='b')
ax.set_yticklabels(y, rotation=0, fontsize=25, color='b')

ax.set_xticks(x)
plt.legend(fontsize=15)
plt.grid(True)
#plt.show()
plt.savefig("images/guineaBissau_nigeria.png")
```
