```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as n
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
burkina = data[(data['home_team'] == 'Burkina Faso') | (data['away_team'] == 'Burkina Faso')]
```

```python
burkina.head()
```

```python
brk_l20 = burkina[-20:]
brk_l20
```

```python
brk_win = 0
brk_lose = 0
brk_draw = 0
for i,j,k,l in zip(brk_l20['home_team'], brk_l20['away_team'], brk_l20['home_score'], brk_l20['away_score']):
    if ((i == 'Burkina Faso') and (k>l)) or ((j == 'Burkina Faso') and (k<l)):
        brk_win +=1
    elif ((i == 'Burkina Faso') and (k<l)) or ((j == 'Burkina Faso') and (k>l)):
        brk_lose +=1
    else:
        brk_draw += 1
    
print(brk_win, brk_lose, brk_draw)
```

```python
ethiopia = data[(data['home_team'] == 'Ethiopia') | (data['away_team'] == 'Ethiopia')]
```

```python
ethiopia.tail()
```

```python
eth_l20 = ethiopia[-20:]
eth_l20
```

```python
eth_win = 0
eth_lose = 0
eth_draw = 0
for i,j,k,l in zip(eth_l20['home_team'], eth_l20['away_team'], eth_l20['home_score'], eth_l20['away_score']):
    if ((i == 'Ethiopia') and (k>l)) or ((j == 'Ethiopia') and (k<l)):
        eth_win +=1
    elif ((i == 'Ethiopia') and (k<l)) or ((j == 'Ethiopia') and (k>l)):
        eth_lose +=1
    else:
        eth_draw += 1
    
print(eth_win, eth_lose, eth_draw)
```

```python
labels = ['WIN', 'LOSE', 'DRAW']

brk = [brk_win, brk_lose, brk_draw]
eth = [eth_win, eth_lose, eth_draw]

width = 0.25

x = np.arange(0, 3)

fig, ax = plt.subplots(figsize=(10,10))

y = list(range(0, 16, 2))

rects1 = ax.bar(x - width/2, brk, width, label='Burkina Faso', color='g')
rects2 = ax.bar(x + width/2, eth, width, label='Ethiopia', color='y')

ax.set_ylabel('Number of Outcomes',fontsize=30, color='r')
ax.set_xlabel("Outcomes", fontsize=30, color='r')
ax.set_title("The last 20 matches of Burkina and Ethiopia", fontsize=25, color='b')
ax.set_xticklabels(labels, fontsize=15, color='b')
ax.set_yticklabels(y, rotation=0, fontsize=25, color='b')

ax.set_xticks(x)
plt.legend(fontsize=15)
plt.grid(True)
plt.savefig("burkina_ethiopia.png")
```
