```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
guinea = data[(data['home_team'] == 'Guinea-Bissau') | (data['away_team'] == 'Guinea-Bissau')]
```

```python
guinea.head()
```

```python
gui_l20 = guinea[-20:]
gui_l20
```

```python
gui_win = 0
gui_lose = 0
gui_draw = 0
for i,j,k,l in zip(gui_l20['home_team'], gui_l20['away_team'], gui_l20['home_score'], gui_l20['away_score']):
    if ((i == 'Guinea-Bissau') and (k>l)) or ((j == 'Guinea-Bissau') and (k<l)):
        gui_win +=1
    elif ((i == 'Guinea-Bissau') and (k<l)) or ((j == 'Guinea-Bissau') and (k>l)):
        gui_lose +=1
    else:
        gui_draw += 1
    
print(gui_win, gui_lose, gui_draw)
```

```python
nigeria = data[(data['home_team'] == 'Nigeria') | (data['away_team'] == 'Nigeria')]
```

```python
nigeria.tail()
```

```python
nig_l20 = nigeria[-20:]
nig_l20
```

```python
nig_win = 0
nig_lose = 0
nig_draw = 0
for i,j,k,l in zip(nig_l20['home_team'], nig_l20['away_team'], nig_l20['home_score'], nig_l20['away_score']):
    if ((i == 'Nigeria') and (k>l)) or ((j == 'Nigeria') and (k<l)):
        nig_win +=1
    elif ((i == 'Nigeria') and (k<l)) or ((j == 'Nigeria') and (k>l)):
        nig_lose +=1
    else:
        nig_draw += 1
    
print(nig_win, gui_lose, nig_draw)
```

```python
labels = ['WIN', 'LOSE', 'DRAW']

gui = [gui_win, gui_lose, gui_draw]
nig = [nig_win, gui_lose, nig_draw]

width = 0.25

x = np.arange(0, 3)

fig, ax = plt.subplots(figsize=(10,10))

y = list(range(0, 16, 2))

rects1 = ax.bar(x - width/2, gui, width, label='Guinea-Bissau', color='r')
rects2 = ax.bar(x + width/2, nig, width, label='Nigeria', color='g')

ax.set_ylabel('Number of Outcomes',fontsize=30, color='r')
ax.set_xlabel("Outcomes", fontsize=30, color='r')
ax.set_title("The last 20 matches of Guinea-Bissau and Nigeria", fontsize=25, color='b')
ax.set_xticklabels(labels, fontsize=15, color='b')
ax.set_yticklabels(y, rotation=0, fontsize=25, color='b')

ax.set_xticks(x)
plt.legend(fontsize=15)
plt.grid(True)
#plt.show()
plt.savefig("images/guineaBissau_nigeria.png")
```
