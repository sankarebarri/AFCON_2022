```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
sierraLeone = data[(data['home_team'] == 'Sierra Leone') | (data['away_team'] == 'Sierra Leone')]
```

```python
sierraLeone.head()
```

```python
sieraL_l20 = sierraLeone[-20:]
sieraL_l20
```

```python
sieL_win = 0
sieL_lose = 0
sieL_draw = 0
for i,j,k,l in zip(sieraL_l20['home_team'], sieraL_l20['away_team'], sieraL_l20['home_score'], sieraL_l20['away_score']):
    if ((i == 'Sierra Leone') and (k>l)) or ((j == 'Sierra Leone') and (k<l)):
        sieL_win +=1
    elif ((i == 'Sierra Leone') and (k<l)) or ((j == 'Sierra Leone') and (k>l)):
        sieL_lose +=1
    else:
        sieL_draw += 1
    
print(sieL_win, sieL_lose, sieL_draw)
```

```python
equatorial = data[(data['home_team'] == 'Equatorial Guinea') | (data['away_team'] == 'Equatorial Guinea')]
```

```python
equatorial.tail()
```

```python
equatorial_l20 = equatorial[-20:]
equatorial_l20
```

```python
equatorial_win = 0
equatorial_lose = 0
equatorial_draw = 0
for i,j,k,l in zip(equatorial_l20['home_team'], equatorial_l20['away_team'], equatorial_l20['home_score'], equatorial_l20['away_score']):
    if ((i == 'Equatorial Guinea') and (k>l)) or ((j == 'Equatorial Guinea') and (k<l)):
        equatorial_win +=1
    elif ((i == 'Equatorial Guinea') and (k<l)) or ((j == 'Equatorial Guinea') and (k>l)):
        equatorial_lose +=1
    else:
        equatorial_draw += 1
    
print(equatorial_win, equatorial_lose, equatorial_draw)
```

```python
labels = ['WIN', 'LOSE', 'DRAW']

sierra = [sieL_win, sieL_lose, sieL_draw]
equito = [equatorial_win, equatorial_lose, equatorial_draw]


width = 0.25

x = np.arange(0, 3)

fig, ax = plt.subplots(figsize=(10,10))

y = list(range(0, 10))

rects1 = ax.bar(x - width/2, sierra, width, label='Sierra Leone', color='g')
rects2 = ax.bar(x + width/2, equito, width, label='Equitorial Guinea', color='r')

ax.set_ylabel('Number of Outcomes',fontsize=30, color='y')
ax.set_xlabel("Outcomes", fontsize=30, color='y')
ax.set_title("The last 20 matches of Sierra Leone and Equitorial Guinea", fontsize=25, color='b')
ax.set_xticklabels(labels, fontsize=15, color='b')
ax.set_yticklabels(y, rotation=0, fontsize=25, color='b')

ax.set_xticks(x)
plt.legend(fontsize=15)
plt.grid(True)
#plt.show()
plt.savefig("images/sierraLeone_equitoGuinea.png")
```
