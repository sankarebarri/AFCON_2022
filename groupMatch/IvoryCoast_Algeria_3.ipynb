```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
ivory = data[(data['home_team'] == 'Ivory Coast') | (data['away_team'] == 'Ivory Coast')]
```

```python
ivory.head()
```

```python
ivory_l20 = ivory[-20:]
ivory_l20
```

```python
ivory_win = 0
ivory_lose = 0
ivory_draw = 0
for i,j,k,l in zip(ivory_l20['home_team'], ivory_l20['away_team'], ivory_l20['home_score'], ivory_l20['away_score']):
    if ((i == 'Ivory Coast') and (k>l)) or ((j == 'Ivory Coast') and (k<l)):
        ivory_win +=1
    elif ((i == 'Ivory Coast') and (k<l)) or ((j == 'Ivory Coast') and (k>l)):
        ivory_lose +=1
    else:
        ivory_draw += 1
    
print(ivory_win, ivory_lose, ivory_draw)
```

```python
algeria = data[(data['home_team'] == 'Algeria') | (data['away_team'] == 'Algeria')]
```

```python
algeria.tail()
```

```python
algeria_l20 = algeria[-20:]
algeria_l20
```

```python
algeria_win = 0
algeria_lose = 0
algeria_draw = 0
for i,j,k,l in zip(algeria_l20['home_team'], algeria_l20['away_team'], algeria_l20['home_score'], algeria_l20['away_score']):
    if ((i == 'Algeria') and (k>l)) or ((j == 'Algeria') and (k<l)):
        algeria_win +=1
    elif ((i == 'Algeria') and (k<l)) or ((j == 'Algeria') and (k>l)):
        algeria_lose +=1
    else:
        algeria_draw += 1
    
print(algeria_win, algeria_lose, algeria_draw)
```

```python
labels = ['WIN', 'LOSE', 'DRAW']

ivo = [ivory_win, ivory_lose, ivory_draw]
alg = [algeria_win, algeria_lose, algeria_draw]


width = 0.25

x = np.arange(0, 3)

fig, ax = plt.subplots(figsize=(10,10))

y = list(range(0, 14, 2))

rects1 = ax.bar(x - width/2, ivo, width, label='Ivory Coast', color='orange')
rects2 = ax.bar(x + width/2, alg, width, label='Algeria', color='g')

ax.set_ylabel('Number of Outcomes',fontsize=30, color='y')
ax.set_xlabel("Outcomes", fontsize=30, color='y')
ax.set_title("The last 20 matches of Ivory Coast and Algeria", fontsize=25, color='b')
ax.set_xticklabels(labels, fontsize=35, color='b')
ax.set_yticklabels(y, rotation=0, fontsize=35, color='b')

ax.set_xticks(x)
plt.legend(fontsize=15)
plt.grid(True)
#plt.show()
plt.savefig("images/ivoryCoast_Algeria.png")
```
