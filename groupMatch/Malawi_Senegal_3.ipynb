```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
malawi = data[(data['home_team'] == 'Malawi') | (data['away_team'] == 'Malawi')]
```

```python
malawi.head()
```

```python
malawi_l20 = malawi[-20:]
malawi_l20
```

```python
malawi_win = 0
malawi_lose = 0
malawi_draw = 0
for i,j,k,l in zip(malawi_l20['home_team'], malawi_l20['away_team'], malawi_l20['home_score'], malawi_l20['away_score']):
    if ((i == 'Malawi') and (k>l)) or ((j == 'Malawi') and (k<l)):
        malawi_win +=1
    elif ((i == 'Malawi') and (k<l)) or ((j == 'Malawi') and (k>l)):
        malawi_lose +=1
    else:
        malawi_draw += 1
    
print(malawi_win, malawi_lose, malawi_draw)
```

```python
senegal = data[(data['home_team'] == 'Senegal') | (data['away_team'] == 'Senegal')]
```

```python
senegal.tail()
```

```python
senegal_l20 = senegal[-20:]
senegal_l20
```

```python
senegal_win = 0
senegal_lose = 0
senegal_draw = 0
for i,j,k,l in zip(senegal_l20['home_team'], senegal_l20['away_team'], senegal_l20['home_score'], senegal_l20['away_score']):
    if ((i == 'Senegal') and (k>l)) or ((j == 'Senegal') and (k<l)):
        senegal_win +=1
    elif ((i == 'Senegal') and (k<l)) or ((j == 'Senegal') and (k>l)):
        senegal_lose +=1
    else:
        senegal_draw += 1
    
print(senegal_win, senegal_lose, senegal_draw)
```

```python
labels = ['WIN', 'LOSE', 'DRAW']

sen = [senegal_win, senegal_lose, senegal_draw]
malwi = [malawi_win, malawi_lose, malawi_draw]

width = 0.25

x = np.arange(0, 3)

fig, ax = plt.subplots(figsize=(10,10))

y = list(range(0, 16, 2))

rects1 = ax.bar(x - width/2, sen, width, label='Senegal', color='g')
rects2 = ax.bar(x + width/2, malwi, width, label='Malawi', color='y')

ax.set_ylabel('Number of Outcomes',fontsize=30, color='r')
ax.set_xlabel("Outcomes", fontsize=30, color='r')
ax.set_title("The last 20 matches of Senegal and Malawi", fontsize=25, color='b')
ax.set_xticklabels(labels, fontsize=15, color='b')
ax.set_yticklabels(y, rotation=0, fontsize=25, color='b')

ax.set_xticks(x)
plt.legend(fontsize=15)
plt.grid(True)
plt.show()
plt.savefig("senegal_malawi.png")
```
