```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
egy_mor = data[((data['home_team'] == 'Egypt') & (data['away_team'] == 'Morocco')) | (data['home_team'] == 'Morocco') & (data['away_team'] == 'Egypt')]
```

```python
egy_mor
```

```python
neutral = egy_mor[egy_mor['neutral']]
neutral
```

```python
not_friendly = egy_mor[egy_mor['tournament'] != 'Friendly']
not_friendly
```

```python
afcon = egy_mor[egy_mor['tournament'] == 'African Cup of Nations']
afcon
```

```python
def get_results(competition, country_A, country_B):
    win_A = 0 # country A win
    draw = 0 # draw of country_A and country_B are equal
    #lost_A = 0 # country A lost
    for i, j, k, l in zip(competition['home_team'], competition['away_team'], competition['home_score'], competition['away_score']):
        if ((i == country_A and k > l) or (j == country_A and k < l)):
            win_A += 1
        elif ((i == country_A and k == l) or (j == country_A and l == k)):
            draw += 1
    return [win_A, draw, len(competition)-(win_A + draw)]
```

```python
all_match = get_results(egy_mor, 'Egypt', 'Morocco')
all_match
```

```python
neu = get_results(neutral, 'Egypt', 'Morocco')
neu
```

```python
not_frnd = get_results(not_friendly, 'Egypt', 'Morocco')
not_frnd
```

```python
can = get_results(afcon, 'Egypt', 'Morocco')
can
```

```python
labels1 = [f'EGYPT WON({neu[0]})', f'DRAW({neu[1]})', f'MOROCCO WON({neu[2]})']
sizes1 = get_results(neutral, 'Egypt', 'Morocco')

labels2 = [f'EGYPT WON({not_frnd[0]})', f'DRAW({not_frnd[1]})', f'MOROCCO WON({not_frnd[2]})']
sizes2 = get_results(not_friendly, 'Egypt', 'Morocco')

labels3 = [f'EGYPT WON({can[0]})', f'DRAW({can[1]})', f'MOROCCO WON({can[2]})']
sizes3 = get_results(afcon, 'Egypt', 'Morocco')

fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1 , figsize=(20,20))

ax1.pie(sizes1, labels=labels1, autopct='%.0f%%')
ax2.pie(sizes2, labels=labels2, autopct='%.0f%%')
ax3.pie(sizes3, labels=labels3, autopct='%.0f%%')


ax1.set_title(f'NEUTRAL MATCHES({sum(neu)})', color='r', fontsize=25)
ax2.set_title(f'NON-FRIENDLY MATCHES({sum(not_frnd)})', color='r', fontsize=25)
ax3.set_title(f'AFRICAN CUP OF NATIONS MATCHES({sum(can)})', color='r', fontsize=25)


#plt.show()
plt.savefig("images/egypt_morocco1.png")
```

```python
labels = [f'EGYPT WON({all_match[0]})', f'DRAW({all_match[1]})', f'MOROCCO WON({all_match[2]})']
sizes = get_results(egy_mor, 'Egypt', 'Morocco')

fig, ax1 = plt.subplots(figsize=(10, 10))

ax1.pie(sizes, labels=labels, autopct='%.0f%%')
ax1.set_title(f'ALL MATCHES({sum(all_match)})', color='r', fontsize=25)


#plt.show()
plt.savefig("images/egypt_morocco2.png")

```
