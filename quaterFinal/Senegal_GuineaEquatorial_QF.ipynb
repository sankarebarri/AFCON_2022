```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
sen_gui = data[((data['home_team'] == 'Senegal') & (data['away_team'] == 'Equatorial Guinea')) | (data['home_team'] == 'Equatorial Guinea') & (data['away_team'] == 'Senegal')]
```

```python
sen_gui
```

```python
labels = ['SENEGAL WON(2)', 'EQUATORIAL-GUINEA WON(1)']
sizes = [2, 1]

fig, ax1 = plt.subplots(figsize=(10, 10))

ax1.pie(sizes, labels=labels, autopct='%.0f%%')
ax1.set_title('ALL MATCHES(3)', color='r', fontsize=25)


#plt.show()
plt.savefig("images/senegal_guinea.png")

```
