```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
ivory_egypt = data[((data['home_team'] == 'Ivory Coast') & (data['away_team'] == 'Egypt')) | (data['home_team'] == 'Egypt') & (data['away_team'] == 'Ivory Coast')]
```

```python
ivory_egypt
```

```python
len(ivory_egypt)
```

```python
neutral = ivory_egypt[ivory_egypt['neutral']]
neutral
```

```python
friendly = ivory_egypt[ivory_egypt['tournament'] == 'Friendly']
friendly
```

```python
not_friendly = ivory_egypt[ivory_egypt['tournament'] != 'Friendly']
not_friendly
```

```python
afcon = ivory_egypt[ivory_egypt['tournament'] == 'African Cup of Nations']
afcon
```

```python
def get_results(competition, country_A, country_B):
    win_A = 0 # country A win
    draw = 0 # draw of country_A and country_B are equal
    #lost_A = 0 # country A lost
    for i, j, k, l in zip(competition['home_team'], competition['away_team'], competition['home_score'], competition['away_score']):
        if ((i == country_A and k > l) or (j == country_A and k < l)):
            win_A += 1
        elif ((i == country_A and k == l) or (j == country_A and k == l)):
            draw += 1
    return [win_A, draw, len(competition)-(win_A + draw)]
```

```python
get_results(ivory_egypt, 'Ivory Coast', 'Egypt')
```

```python
labels1 = ['IVORY COAST WON(1)', 'DRAW(2)', 'EGYPT WON(7)']
sizes1 = get_results(afcon, 'Ivory Coast', 'Egypt')

labels2 = ['IVORY COAST WON(4)', 'DRAW(3)', 'EGYPT WON(1)']
sizes2 = get_results(friendly, 'Ivory Coast', 'Egypt')

labels3 = ['IVORY COAST WON(3)', 'DRAW(3)', 'EGYPT WON(8)']
sizes3 = get_results(not_friendly, 'Ivory Coast', 'Egypt')

labels4 = ['IVORY COAST WON(3)', 'DRAW(2)', 'EGYPT WON(3)']
sizes4 = get_results(neutral, 'Ivory Coast', 'Egypt')

fig, ([ax1, ax2], [ax3, ax4]) = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

ax1.pie(sizes1, labels=labels1, autopct='%.0f%%')
ax2.pie(sizes2, labels=labels2, autopct='%.0f%%')
ax3.pie(sizes3, labels=labels3, autopct='%.0f%%')
ax4.pie(sizes4, labels=labels4, autopct='%.0f%%')

ax1.set_title('AFCON MATCHES(7)', color='b', fontsize='15')
ax2.set_title('FRIENDLY MATCHES(8)', color='b', fontsize='15')
ax3.set_title('NON-FRIENDLY MATCHES(14)', color='b', fontsize='15')
ax4.set_title('NEUTRAL GROUNDS(8)', color='b', fontsize='15')

#plt.show()
plt.savefig("images/ivoryCoast_egypt1.png")
```
