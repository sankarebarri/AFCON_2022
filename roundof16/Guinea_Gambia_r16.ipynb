```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
guinea_gambie = data[((data['home_team'] == 'Guinea') & (data['away_team'] == 'Gambia')) | (data['home_team'] == 'Gambia') & (data['away_team'] == 'Guinea')]
```

```python
neutral = guinea_gambie[guinea_gambie['neutral']]
neutral
```

```python
friendly = guinea_gambie[guinea_gambie['tournament'] == 'Friendly']
friendly
```

```python
not_friendly = guinea_gambie[guinea_gambie['tournament'] != 'Friendly']
not_friendly
```

```python
afcon = guinea_gambie[guinea_gambie['tournament'] == 'African Cup of Nations']
afcon
```

```python
def get_results(competition, country_A, country_B):
    win_A = 0 # country A win
    draw = 0 # draw of country_A and country_B are equal
    #lost_A = 0 # country A lost
    for i, j, k, l in zip(competition['home_team'], competition['away_team'], competition['home_score'], competition['away_score']):
        if ((i == country_A and k > l) or (j == country_A and k < l)):
            win_A += 1
        elif ((i == country_A and k == l) or (j == country_A and k == l)):
            draw += 1
    return [win_A, draw, len(competition)-(win_A + draw)]
```

```python
get_results(neutral, 'Guinea', 'Gambia')
```

```python
labels1 = ['GUINEA WON', 'DRAW', 'GAMBIA WON']
sizes1 = get_results(neutral, 'Guinea', 'Gambia')

labels2 = ['GUINEA WON', 'DRAW', 'GAMBIA WON']
sizes2 = get_results(friendly, 'Guinea', 'Gambia')

labels3 = ['GUINEA WON', 'DRAW', 'GAMBIA WON']
sizes3 = get_results(not_friendly, 'Guinea', 'Gambia')


fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(10, 10))

ax1.pie(sizes1, labels=labels1, autopct='%.0f%%')
ax2.pie(sizes2, labels=labels2, autopct='%.0f%%')
ax3.pie(sizes3, labels=labels3, autopct='%.0f%%')


ax1.set_title('NEUTRAL GROUNDS', fontsize=25, color='b')
ax2.set_title('FRIENDLY MATCHES', fontsize=25, color='b')
ax3.set_title('NON FRIENDLY MATCHES', fontsize=25, color='b')


#plt.show()
plt.savefig("images/guinea_gambia1.png")
```

```python
labels1 = ['GUINEA WON', 'DRAW', 'GAMBIA WON']
sizes1 = get_results(guinea_gambie, 'Guinea', 'Gambia')

fig, ax1 = plt.subplots( figsize=(10, 10))

ax1.pie(sizes1, labels=labels1, autopct='%.0f%%')



ax1.set_title('ALL MATCHES', fontsize=25, color='b')

#plt.show()
plt.savefig("images/guinea_gambia2.png")
```
