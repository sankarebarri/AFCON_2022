```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
bur_gabon = data[((data['home_team'] == 'Burkina Faso') & (data['away_team'] == 'Gabon')) | (data['home_team'] == 'Gabon') & (data['away_team'] == 'Burkina Faso')]
```

```python
len(bur_gabon)
```

```python
neutral = bur_gabon[bur_gabon['neutral']]
neutral
```

```python
friendly = bur_gabon[bur_gabon['tournament'] == 'Friendly']
friendly
```

```python
not_friendly = bur_gabon[bur_gabon['tournament'] != 'Friendly']
not_friendly
```

```python
afcon = bur_gabon[bur_gabon['tournament'] == 'African Cup of Nations']
afcon
```

```python
def get_results(competition, country_A, country_B):
    win_A = 0 # country A win
    draw = 0 # draw of country_A and country_B are equal
    #lost_A = 0 # country A lost
    for i, j, k, l in zip(competition['home_team'], competition['away_team'], competition['home_score'], competition['away_score']):
        if ((i == country_A and k > l) or (j == country_A and k < l)):
            win_A += 1
        elif ((i == country_A and k == l) or (j == country_A and l == k)):
            draw += 1
    return [win_A, draw, len(competition)-(win_A + draw)]
```

```python
get_results(afcon, 'Burkina Faso', 'Gabon')
```

```python
labels1 = ['BURKINA WON', 'DRAW', 'GABON WON']
sizes1 = get_results(afcon, 'Burkina Faso', 'Gabon')

labels2 = ['BURKINA WON', 'DRAW', 'GABON WON']
sizes2 = get_results(friendly, 'Burkina Faso', 'Gabon')

labels3 = ['BURKINA WON', 'DRAW', 'GABON WON']
sizes3 = get_results(not_friendly, 'Burkina Faso', 'Gabon')

labels4 = ['BURKINA WON', 'DRAW', 'GABON WON']
sizes4 = get_results(neutral, 'Burkina Faso', 'Gabon')

fig, ([ax1, ax2], [ax3, ax4]) = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

ax1.pie(sizes1, labels=labels1, autopct='%.0f%%')
ax2.pie(sizes2, labels=labels2, autopct='%.0f%%')
ax3.pie(sizes3, labels=labels3, autopct='%.0f%%')
ax4.pie(sizes4, labels=labels4, autopct='%.0f%%')

ax1.set_title('AFCON MATCHES')
ax2.set_title('FRIENDLY MATCHES')
ax3.set_title('NON-FRIENDLY MATCHES')
ax4.set_title('NEUTRAL GROUNDS')

plt.show()
```

```python
labels = ['BURKINA WON', 'DRAW', 'GABON WON']
sizes = get_results(bur_gabon, 'Burkina Faso', 'Gabon')

fig, ax1 = plt.subplots(figsize=(10, 10))

ax1.pie(sizes, labels=labels, autopct='%.0f%%')
ax1.set_title('BURKINA vs GABON', fontsize=25)


plt.show()
```
