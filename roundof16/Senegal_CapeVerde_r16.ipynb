```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/results.csv')
data.head()
```

```python
sen_cape = data[((data['home_team'] == 'Senegal') & (data['away_team'] == 'Cape Verde')) | (data['home_team'] == 'Cape Verde') & (data['away_team'] == 'Senegal')]
```

```python
def get_results(competition, country_A, country_B):
    win_A = 0 # country A win
    draw = 0 # draw of country_A and country_B are equal
    #lost_A = 0 # country A lost
    for i, j, k, l in zip(competition['home_team'], competition['away_team'], competition['home_score'], competition['away_score']):
        if ((i == country_A and k > l) or (j == country_A and k < l)):
            win_A += 1
        elif ((i == country_A and k == l) or (j == country_A and k == l)):
            draw += 1
    return [win_A, draw, len(competition)-(win_A + draw)]
```

```python
get_results(friendly, 'Senegal', 'Cape Verde')
```

```python
labels1 = ['SENEGAL WON(6)', 'DRAW(0)', 'CAPE_VERDE WON(1)']
sizes1 = get_results(neutral, 'Senegal', 'Cape Verde')

labels2 = ['SENEGAL WON(6)', 'CAPE_VERDE WON(0)']
sizes2 = [6, 0]

labels3 = ['SENEGAL WON(9)', 'DRAW(2)', 'CAPE_VERDE WON(2)']
sizes3 = get_results(not_friendly, 'Senegal', 'Cape Verde')


fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(10, 10))

ax1.pie(sizes1, labels=labels1, autopct='%.0f%%')
ax2.pie(sizes2, labels=labels2, autopct='%.0f%%')
ax3.pie(sizes3, labels=labels3, autopct='%.0f%%')


ax1.set_title('NEUTRAL GROUND MATCHES(7)', fontsize=25, color='b')
ax2.set_title('FRIENDLY MATCHES(6)', fontsize=25, color='b')
ax3.set_title('NON FRIENDLY MATCHES(13)', fontsize=25, color='b')


#plt.show()
plt.savefig("images/senegal_capeverde1.png")
```
