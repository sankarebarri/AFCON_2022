```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
countries = ['Cameroon', 'Burkina Faso', 'Cape Verde', 'Ethiopia', 'Senegal', 'Guinea', 'Malawi', 'Zimbabwe',
            'Morocco', 'Gabon', 'Comoros', 'Ghana', 'Nigeria', 'Egypt', 'Sudan', 'Guinea-Bissau',
            'Ivory Coast', 'Equatorial Guinea', 'Sierra Leone', 'Algeria', 'Mali', 'Gambia', 'Tunisia', 'Mauritania']
```

```python
len(countries)
```

```python
goals_by_country = [14, 9, 2, 2, 9, 2, 3, 3, 8, 5, 4, 3, 6, 4, 1, 0, 6, 3, 2, 1, 4, 4 , 5, 0]
```

```python
len(goals_by_country)
```

```python
plt.tight_layout()
plt.figure(figsize=(9,10))
plt.xlabel('Number of goals', fontsize=25, color='b')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20, color='b')
plt.title('Number of goals scored by each country in the AFCON', fontsize=20, color='r')
plt.barh(countries, goals_by_country)
plt.savefig("images/country_goals.png")
#plt.savefig('country_goals', bbox_inches = 'tight')
```
